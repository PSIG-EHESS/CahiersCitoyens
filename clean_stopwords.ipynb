```python
import pandas as pd
import os
import os

list_dir = ["output_lemmas", "output_pos"]
for dir_path in list_dir :
    count = 0
    for path in os.listdir(dir_path):
        # check if current path is a file
        count += 1
    print(f"dossier {dir_path} : {count} fichiers")
```

```python
s_lemma = pd.Series(dtype = str)
s_pos = pd.Series(dtype = str)

# go through all the files saved after lemmatization
for i in range(3) :
    one_concat_lemma = pd.read_csv(f"output_lemmas/lemma_temp_{i}.csv", sep = ",", dtype = str)
    one_concat_pos = pd.read_csv(f"output_pos/pos_temp_{i}.csv", sep = ",", dtype = str)
    s_lemma = pd.concat([s_lemma,one_concat_lemma], axis = 0, ignore_index=True)
    s_pos = pd.concat([s_pos,one_concat_pos], axis = 0, ignore_index=True)


# transpose each serie as read_csv read it in the other axis
s_lemma = s_lemma.T
s_pos = s_pos.T
# remove first row which is a 0
s_lemma = s_lemma.drop([0])
s_pos = s_pos.drop([0])
s_lemma
```

```python
from spacy.lang.fr.stop_words import STOP_WORDS as fr_stop
from spacy.lang.en.stop_words import STOP_WORDS as en_stop

final_stopwords_list = list(fr_stop) + list(en_stop) + ["\n"]
final_stopwords_set = set(final_stopwords_list)

def remove_stopwords (text):
    print(text)
    buff =[]
    for word in text :
        if word not in final_stopwords_list :
            buff.append(word)
    output = " ".join(buff)
#     text = [word for word in text if word not in final_stopwords_set]
    return output
```

```python
p_rem = lambda x : remove_stopwords(x)
t = s_lemma.apply(p_rem)
print(t)
```
